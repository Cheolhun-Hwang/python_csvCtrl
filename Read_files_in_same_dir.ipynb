```python
for i in Ship_list:
    df_ship = df2.loc[df2['MMSI'] == i]
    depart = pd.to_numeric(df_ship.loc[df_ship.index.min(), 'LONG':'LAT'])
    end = pd.to_numeric(df_ship.loc[df_ship.index.max(), 'LONG':'LAT'])
    Distance = haversine(depart, end)
    print('Distance: ', Distance, 'Km')
```
